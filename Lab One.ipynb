```python
#Bin the Data and find the mean for each segment
lowav = np.array([])
av25 = np.array([])
midav = np.array([])
av75 = np.array([])
highav = np.array([])
darkav = np.array([])
N = 1000
a = np.arange(dt_low.size)
b = np.arange(dt_m_l_25.size)
c = np.arange(dt_mid.size)
d = np.arange(dt_m_h_75.size)
e = np.arange(dt_high.size)
f = np.arange(dt_dark.size)
for i in a[0::N]:
    lowmean = np.mean(dt_low[0:i+N])
    lowav = np.append(lowav, lowmean)
    print('mean of low from 0 to ', i, '= ', lowmean)
for i in b[0::N]:
    mean25 = np.mean(dt_m_l_25[0:i+N])
    av25 = np.append(av25, mean25)
    print('mean of 25% from 0 to ', i, '= ', mean25)
for i in c[0::N]:
    midmean = np.mean(dt_mid[0:i+N])
    midav = np.append(midav, midmean)
    print('mean of mid from 0 to ', i, '= ', midmean)
for i in d[0::N]:
    mean75 = np.mean(dt_m_h_75[0:i+N])
    av75 = np.append(av75, mean75)
    print('mean of 75% from 0 to ', i, '= ', mean75)
for i in e[0::N]:
    highmean = np.mean(dt_high[0:i+N])
    highav = np.append(highav, highmean)
    print('mean of high from 0 to ', i, '= ', highmean)
for i in f[0::N]:
    darkmean = np.mean(dt_dark[0:i+N])
    darkav = np.append(darkav, darkmean)
    print('mean of dark from 0 to ', i, '= ', darkmean)
```

```python
#Plot the binned means
txt = "x-axis indices correspond to the dt array indices by a factor of {}"
plot_lowav = plt.plot(lowav, 'o', label='low', color = 'b')
plt.xlabel('Start Index' )
plt.ylabel('Average')
plt.legend()
plt.figtext(0.1,-0.02,txt.format(N))
plt.show()
plot_av25 = plt.plot(av25, 'o', label = '25%', color ='r')
plt.xlabel('Start Index (Thousand)')
plt.ylabel('Average')
plt.legend()
plt.figtext(0.1,-0.02,txt.format(N))
plt.show()
plot_midav = plt.plot(midav, 'o', label = 'mid', color = 'g')
plt.xlabel('Start Index (Thousand)')
plt.ylabel('Average')
plt.legend()
plt.figtext(0.1,-0.02,txt.format(N))
plt.show()
plot_av75 = plt.plot(av75, 'o', label = '75%', color = 'y')
plt.xlabel('Start Index (Thousand)')
plt.ylabel('Average')
plt.legend()
plt.figtext(0.1,-0.02,txt.format(N))
plt.show()
plot_highav = plt.plot(highav, 'o', label = 'high', color ='m')
plt.xlabel('Start Index (Thousand)')
plt.ylabel('Average')
plt.legend()
plt.figtext(0.1,-0.02,txt.format(N))
plt.show()
plot_darkav = plt.plot(darkav, 'o', label = 'dark', color = 'c')
plt.xlabel('Start Index (Thousand)')
plt.ylabel('Average')
plt.legend()
plt.figtext(0.1,-0.02,txt.format(N))
plt.show()
```

```python
#Work out Standard Deviation
mu_low = np.sum(lowav)/np.float(lowav.size*N)
mu_lm25 = np.sum(av25)/np.float(av25.size*N)
mu_mid = np.sum(midav)/np.float(midav.size*N)
mu_mh75 = np.sum(av75)/np.float(av75.size*N)
mu_high = np.sum(highav)/np.float(highav.size*N)
mu_dark = np.sum(darkav)/np.float(darkav.size*N)
s_low = np.array([])
s_lm25 = np.array([])
s_mid = np.array([])
s_mh75 = np.array([])
s_high = np.array([])
s_dark = np.array([])
for i in range(0,len(lowav)):
    s = np.sqrt(np.sum((lowav[i]-mu_low)**2)/(np.float(lowav.size*N)-1))
    s_low = np.append(s_low,s)
for i in range(0,len(av25)):
    s = np.sqrt(np.sum((av25[i]-mu_lm25)**2)/(np.float(av25.size*N)-1))
    s_lm25 = np.append(s_lm25,s)
for i in range(0,len(midav)):
    s = np.sqrt(np.sum((midav[i]-mu_mid)**2)/(np.float(midav.size*N)-1))
    s_mid = np.append(s_mid,s)
for i in range(0,len(av75)):
    s = np.sqrt(np.sum((av75[i]-mu_mh75)**2)/(np.float(av75.size*N)-1))
    s_mh75 = np.append(s_mh75,s)
for i in range(0,len(highav)):
    s = np.sqrt(np.sum((highav[i]-mu_high)**2)/(np.float(highav.size*N)-1))
    s_high = np.append(s_high,s)
for i in range(0,len(darkav)):
    s = np.sqrt(np.sum((darkav[i]-mu_dark)**2)/(np.float(darkav.size*N)-1))
    s_dark = np.append(s_dark,s)
```

```python
#Standard Deviation of Mean?

d_low = s_low/np.sqrt(lowav.size)

d_lm25 = s_lm25/np.sqrt(N)

d_mid = s_mid/np.sqrt(N)

d_mh75 = s_mid/np.sqrt(N)

d_high = s_high/np.sqrt(N)

d_dark = s_dark/np.sqrt(N)

ncount = np.linspace(1,sample,N)
ones = np.ones(s_low.size)
d_ther = 1/np.sqrt(ncount)
```
